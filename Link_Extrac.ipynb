```python
import os
import pandas as pd

# --- CONFIGURACIÓN ---
# Reemplaza con la ruta de tu carpeta local de GitHub
ruta_fotos = r'C:\GitHub\Fotos' 
usuario_github = 'edwcru'  # Tu usuario de ArcGIS/GitHub
repositorio = 'Fotos'
rama = 'main'

# --- PROCESO ---
lista_datos = []

# Listar archivos en la carpeta
for archivo in os.listdir(ruta_fotos):
    # Filtramos para que solo tome imágenes comunes
    if archivo.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.tiff')):
        # Generar el link web (blob)
        link_web = f"https://github.com/{usuario_github}/{repositorio}/blob/{rama}/{archivo}"
        
        # Guardar nombre y link en una lista
        lista_datos.append({
            "Nombre Real": archivo,
            "Link Web": link_web
        })

# Crear DataFrame y exportar a Excel
df = pd.DataFrame(lista_datos)
df.to_excel("Lista_Fotos_GitHub.xlsx", index=False)

print(f"¡Listo! Se ha generado el archivo 'Lista_Fotos_GitHub.xlsx' con {len(df)} fotos.")
```
